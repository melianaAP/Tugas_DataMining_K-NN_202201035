Impor Perpustakaan yang digunakan

```python
import numpy as np # mengelola data numerik
import matplotlib.pyplot as plt # untuk membuat grafik
import pandas as pd # mengelola data excel
```

Impor Dataset

```python
dataset = pd.read_csv('Customer_Behaviour.csv') # memanggil data set
X = dataset.iloc[:, [2,3]].values # membagi nilai x untuk ambil yang berisi kolom
y = dataset.iloc[:, -1].values # membagi nilai y dari kolom terakhir
```

Membagi Dataset

```python
#membagi data menjadi data traning dan data test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

Skala Fitur

```python
from sklearn.preprocessing import StandardScaler # skala untuk selesih data x
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print (X_train)
```

```python
print (X_test)
```

Latihan model K-NN dengan data training

```python
# model K-NN
from sklearn.neighbors import KNeighborsClassifier # memanggil model K-NN
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2) # memanggil model K-NN
classifier.fit(X_train, y_train) # memanggil model
```

Memprediksi Hasil Set Uji

```python
#menghitung nilai prediksi
y_pred = classifier.predict(X_test)
```

```python
# mengtung keakurasian hasil prediksi
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score # melakukan evaluasi dengan accuracy score
print(accuracy_score(y_test, y_pred)) # menampilkan accuracy score dengan data testing dan data prediksi
print ("tingkat akurasi : ", accuracy_score(y_test, y_pred)*100, "%") # menampilkan tingkat akurasi dengan data testing dan data prediksi
```

Pelatihan visualisasi data

```python
from matplotlib.colors import ListedColormap # melakukan visualisasi data dengan matplotlib
X_set, y_set = X_train, y_train # melakukan visualisasi data dengan matplotlib dengan data training
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01), # melakukan visualisasi data dengan matplotlib dengan data training
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01)) # melakukan visualisasi data dengan matplotlib dengan data training
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), # melakukan visualisasi data dengan matplotlib dengan data training
             alpha = 0.75, cmap = ListedColormap(('red', 'green'))) # melakukan visualisasi data dengan matplotlib dengan data training
plt.xlim(X1.min(), X1.max()) # melakukan visualisasi data dengan matplotlib dengan data training
plt.ylim(X2.min(), X2.max()) # melakukan visualisasi data dengan matplotlib dengan data training
for i, j in enumerate(np.unique(y_set)): # melakukan visualisasi data dengan matplotlib dengan data training
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], # melakukan visualisasi data dengan matplotlib
                c = ListedColormap(('blue', 'black'))(i), label = j) # melakukan visualisasi data dengan matplotlib
plt.title('klasifikasi data menggunakan Algoritma K-NN (Data Training)') # melakukan visualisasi data dengan matplotlib
plt.xlabel('umur') # melakukan visualisasi data dengan matplotlib
plt.ylabel('estimasi gaji') # melakukan visualisasi data dengan matplotlib
plt.legend() # melakukan visualisasi data dengan matplotlib
plt.show() # menampilkan visualisasi data dengan matplotlib
```
